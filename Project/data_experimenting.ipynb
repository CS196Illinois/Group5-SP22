```python
import os
import json
```

```python
json_file = json.load(open('data/ece310.json'))
json_file
```

```python
sentences = []
sentence = []

for itm in json_file:
    sentence += [itm['w']]
    if itm['w'][-1] == '.' or itm['w'][-1] == '?': # check for end of sentence -> the [-1] means look at last char
        sentences += [' '.join(sentence)]
        sentence = []
        
sentences
```

```python
tmp_json = json_file
```

```python
start_time = json_file[0]['s']
for itm in tmp_json:
    itm['s'] -= start_time
    itm['e'] -= start_time

tmp_json
```

```python
for itm in tmp_json:
    itm['s'] = round(.001 * itm['s'])
    itm['e'] = round(.001 * itm['e'])

tmp_json
```

```python
with open('data/exp.json', 'w') as f:
    json.dump(tmp_json, f)
```
